# **Project Protfolio -  Insurance Cost Prediction**

# **Problem Statement**

- Insurance companies need to accurately predict the cost of health insurance for individuals to set premiums appropriately.
- However, traditional methods of cost prediction often rely on broad actuarial tables and historical averages, which may not account for the nuanced differences among individuals.
- By leveraging machine learning techniques, insurers can predict more accurately the insurance costs tailored to individual profiles, leading to more competitive pricing and better risk management.


# **Insurance Cost Prediction Need**

The primary need for this project arises from the challenges insurers face in pricing policies accurately while remaining competitive in the market. Inaccurate predictions can lead to losses for insurers and unfairly high premiums for policyholders. By implementing a machine learning model, insurers can:

- **Enhance Precision in Pricing :** Use individual data points to determine premiums that reflect actual risk more closely than generic estimates.
- **Increase Competitiveness :** Offer rates that are attractive to consumers while ensuring that the pricing is sustainable for the insurer.
- **Improve Customer Satisfaction :** Fair and transparent pricing based on personal health data can increase trust and satisfaction among policyholders.
- **Enable Personalized Offerings :** Create customized insurance packages based on predicted costs, which can cater more directly to the needs and preferences of individuals.
- **Risk Assessment :** Insurers can use the model to refine their risk assessment processes, identifying key factors that influence costs most significantly.
- **Policy Development :** The insights gained from the model can inform the development of new insurance products or adjustments to existing ones.
- **Strategic Decision Making :** Predictive analytics can aid in broader strategic decisions, such as entering new markets or adjusting policy terms based on risk predictions.
- **Customer Engagement :** Insights from the model can be used in customer engagement initiatives, such as personalized marketing and tailored advice for policyholders.

# **FEATURE ENGINEERING**


```python
# To implement this logic for categorizing age groups a new feature with the help of age columns

# Define a function for categorizing age
def categorize_age(age):
    if 18 <= age <= 24:
        return "Young Adult"
    elif 25 <= age <= 34:
        return "Adult"
    elif 35 <= age <= 44:
        return "Mid-Age Adult"
    elif 45 <= age <= 54:
        return "Middle-Aged"
    elif 55 <= age <= 64:
        return "Pre Seniors"
    else:
        return "Seniors"

# Apply the function to create a new column
df['AgeGroup'] = df['Age'].apply(categorize_age)

# Display the first few rows to check the new column
print(df[['Age', 'AgeGroup']].head())
```

```python
# To create a new feature to analysis the health status of the person

def health_status(row):
    if (row['Diabetes'] == 1 or
        row['AnyTransplants'] == 1 or
        row['AnyChronicDiseases'] == 1 or
        row['BloodPressureProblems'] == 1 or
        row['HistoryOfCancerInFamily'] == 1 or
        row['KnownAllergies'] == 1 or
        row['NumberOfMajorSurgeries'] in [1, 2, 3]):
        return "UnHealthy"
    else:
        return "Healthy"

# Apply the function to create a new column
df['HealthStatus'] = df.apply(health_status, axis=1)
```

```python
# Calaculate BMI for each person

df['BMI'] = df['Weight'] / (df['Height'] / 100)** 2
```

```python
# BMI category feature

def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'

df['BMICategory'] = df['BMI'].apply(bmi_category)
```

# **HYPOTHESIS TESTING**

**FRAMEWORK FOR HYPOTHESIS TESTING**

 Check Basis Assumptions of Hypothesis
- **STEP 1 :** Set up **Null and Alternate Hypothesis**.
- **STEP 2 :** Decide the **Test Statistics** and their **Corresponding Distribution**.
- **STEP 3 :** Test should be **Left Tailed** / **Right Tailed** / **Two Tailed**.
- **STEP 4 :** Determine the **P-Value**.
- **STEP 5 :** Choose **Significance level** (Alpha) and Comparing the obtain P-Value with the alpha.
   *   if **P-Value < alpha** ---> **Reject the Null Hypothesis(H0)**
   *   if **P-Value > alpha** ---> **Accept/Failed to Reject the Null Hypothesis (H0)**


```python
df.head()
```

## **T-TEST**

**STEP-1 : Set up Null Hypothesis**

 - **Null Hypothesis ( H0 )** - There is **no significant** difference in the **PremiumPrice** between individuals **with and without diabetes**.

- **Alternate Hypothesis ( HA )** - There is a **significant** difference in the **PremiumPrice** between individuals **with and without diabetes**.

---
**STEP-2 :** Checking for Basic Assumpitons for the hypothesis

- Distribution check using **QQ Plot**

- Homogeneity of Variances using **Lavene's test**

---
**STEP-3: Define Test statistics; Distribution of T under H0.**

- If the assumptions of T Test are met then we can proceed performing T Test for independent samples else we will perform the non parametric test equivalent to T Test for independent sample i.e., Mann-Whitney U rank test for two independent samples.


---


**STEP-4: Compute the p-value and fix value of alpha.**

- We set our **alpha = 0.05**


---


**STEP-5: Compare p-value and alpha.**

Based on p-value, we will reject or fail to reject H0.

- if **P-Value < alpha** ---> **Reject the Null Hypothesis(H0)**
- if **P-Value > alpha** ---> **Accept/Failed to Reject the Null Hypothesis (H0)**



### 1. **Hypothesis Testing for Premium Price and Diabetes Status.**

**Checking for Basic Assumpitonsof hypothesis Testing for T-TEST**

**STEP 2:** **Distribution check using QQ Plot**

```python

import matplotlib.pyplot as plt
import scipy.stats as sm
import statsmodels.api as sm

# Create a figure with subplots
plt.figure(figsize=(15, 5))

# First subplot for Diabetes
plt.subplot(1, 2, 1)
sm.qqplot(df['Diabetes'], line='s', ax=plt.gca())
plt.title('QQ Plot for Diabetes')

# Second subplot for premium_price
plt.subplot(1, 2, 2)
sm.qqplot(df['PremiumPrice'], line='s', ax=plt.gca())
plt.title('QQ Plot for Premium Price')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.suptitle('QQ Plots for Diabetes and Premium Price', fontsize=16, y=1.02)
plt.show()
```

**INFERENCE:**
 The Above **QQ Plot** indicates that the **data is Not Normally Distributed**. Therefore, we will use the **Shapiro-Wilk Test** as an alternative method to **check Normality**.

**SHAPIRO-WILK'S TEST**

```python
# SHAPIRO-WILK'S TEST FOR NORMALITY -

# H0 : Data is Gaussian
# Ha : Data is Not Gaussian

from scipy.stats import shapiro

# Perform Shapiro-Wilk test for Diabetes
test_stat, p_value = shapiro(df['Diabetes'])
print('p-value', p_value)
print('Test Statistic:', test_stat)

# For alpha =0.05 i.e., 95% confidence level
alpha = 0.05

if p_value < alpha:
  print("Reject HO, Data is Not Gaussian for Diabetes")
else:
  print("Fail to Reject HO, Data is Gaussian for Diabetes")
```

```python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test for Premium Price
test_stat, p_value = shapiro(df['PremiumPrice'])
print('p-value', p_value)
print('Test Statistic:', test_stat)

# For alpha =0.05 i.e., 95% confidence level
alpha = 0.05

if p_value < alpha:
  print("Reject HO, Data is Not Gaussian for Premium Price")
else:
  print("Fail to Reject HO, Data is Gaussian for Premium Price")
```

**INFERENCE:**

- From **QQ Plots**, **Shapiro-Wilk's Test** it is clear that the **Data is Not Normal**.

- Let us check for **Equal Variance** in Each Group- **Levene's Test**

**LEVENE'S TEST**

```python
# LEVENE'S TEST CHECK FOR EQUAL VARIANCE

# H0 - Homogenous Variance (  Equal Variance)
# HA - Non Homogenous Variance (Not Equal Variance)

from scipy.stats import levene

# Perform Levene's test for equal variance
test_stat, p_value = levene(df['Diabetes'], df['PremiumPrice'])

# Print test results
print('p-value:', p_value)
print('Test Statistic:', test_stat)

# Define significance level
alpha = 0.05

# Check result
if p_value < alpha:
    print('REJECT H0 - The Samples do not have Homogeneous Variance')
else:
    print('FAIL TO REJECT H0 - The Samples have Homogeneous Variance')
```

**INFERENCE:**

- **Levene's Test** indicates that the assumption of **Equal Variance is not met**.

- The **assumptions** required for **the T-Test** are **Not Satisfied**.


Since the **samples** are **Not Normally** Distributed, the **T-Test is not applicable** here. Instead, we can use its **non-parametric** equivalent, the **Mann-Whitney U Test** for Independent t-test (two independent groups).


### **Mann-Whitney U Test**

```python
# Mann-Whitney U Test

# H0 - Diabetes and Premium Price are Similar
# Ha - Diabetes and Premium Price are Different


from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U Test
test_stat, p_value = mannwhitneyu(df['Diabetes'], df['PremiumPrice'], alternative='two-sided')

# Print test results
print('P-value:', p_value)
print('Test Statistic:', test_stat)

# Define significance level
alpha = 0.05

# Check result
if p_value < alpha:
    print('REJECT H0 - Diabetes and Premium Price are Different')
else:
    print('FAIL TO REJECT H0 - Diabetes and Premium Price are Similar')
```

**INFERENCE:**

Here **p_value is less than alpha**. So, we **Reject Null Hypothesis**. Hence the **Mann-Whitney U Test** also **confirms** that **Diabetes and Premium Price** are **Different**.

### **2. Hypothesis Testing For Health Condition and Premium Price**


- **Null Hypothesis (H):** There is **No Significant difference** between **Health Condition** and **Premium Prices**.

- **Alternative Hypothesis (H):** There is **Significant difference** between **Health Condition** and **Premium Prices**.

```python
df.groupby('HealthStatus')['PremiumPrice'].mean()
```

```python
from scipy.stats import ttest_ind, shapiro, levene

# Checking normality
print("Shapiro-Wilk Test for Normality:")
print("Healthy:", shapiro(df[df['HealthStatus'] == 'Healthy']['PremiumPrice']).pvalue)
print("UnHealthy:", shapiro(df[df['HealthStatus'] == 'UnHealthy']['PremiumPrice']).pvalue)

# Checking homogeneity of variance
print("\nLevenes Test for Equal Variance:")
print(levene(df[df['HealthStatus'] == 'Healthy']['PremiumPrice'],
             df[df['HealthStatus'] == 'UnHealthy']['PremiumPrice']).pvalue)

# Performing T-test
t_stat, p_value = ttest_ind(df[df['HealthStatus'] == 'Healthy']['PremiumPrice'],
                            df[df['HealthStatus'] == 'UnHealthy']['PremiumPrice'],
                            equal_var=False)  # Welchs T-test if variances are unequal

print("\nT-test Results:")
print('Test Statistic:', t_stat)
print('P-value:', p_value)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject H: Health Status significantly affects Premium Price.")
else:
    print("Fail to Reject H: No significant difference in Premium price based on Health Status.")
```

- Since the assumptions for the T-test (normality and homogeneity of variance) **have failed** because of P- value is very less when compared to alpha , we should use a **non-parametric test**  the **Mann-Whitney U Test** should perfered for independent T- test.

### **Mann-Whitney U Test**

```python
# Mann-Whitney U Test

# H0 : No Significant difference in Premium price based on Health Status.
# HA: Health Status significantly affects Premium Price.

from scipy.stats import mannwhitneyu

# Extracting premium prices for each health status group
group1 = df[df['HealthStatus'] == 'Healthy']['PremiumPrice']
group2 = df[df['HealthStatus'] == 'UnHealthy']['PremiumPrice']

# Perform Mann-Whitney U Test
stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')

print("Mann-Whitney U Test Results:")
print('Test Statistic:', stat)
print('P-value:', p_value)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject H: Health Status significantly affects Premium Price.")
else:
    print("Fail to Reject H: No significant difference in Premium price based on Health Status.")
```

**SHAPIRO-WILK'S TEST**

```python
# H0 : Data is Gaussian
# Ha : Data is Not Gaussian

from scipy.stats import shapiro

# Grouping data based on the number of major surgeries
group1 = df[df['NumberOfMajorSurgeries'] == 0]['PremiumPrice']
group2 = df[df['NumberOfMajorSurgeries'] == 1]['PremiumPrice']
group3 = df[df['NumberOfMajorSurgeries'] == 2]['PremiumPrice']
group4 = df[df['NumberOfMajorSurgeries'] == 3]['PremiumPrice']

# Perform Shapiro-Wilk test for each group
def shapiro_test(group, group_name):
    test_stat, p_value = shapiro(group)
    print(f"Shapiro-Wilk Test for {group_name}:")
    print(f"P-value: {p_value}")
    print(f"Test Statistic: {test_stat}")

    # Alpha level
    alpha = 0.05
    if p_value < alpha:
        print(f"Reject H0 - {group_name}: Data is Not Gaussian")
    else:
        print(f"Fail to Reject H0 - {group_name}: Data is Gaussian")
    print()

# Apply the Shapiro-Wilk test to each group
shapiro_test(group1, 'Group 1 (NumberOfMajorSurgeries == 0)')
shapiro_test(group2, 'Group 2 (NumberOfMajorSurgeries == 1)')
shapiro_test(group3, 'Group 3 (NumberOfMajorSurgeries == 2)')
shapiro_test(group4, 'Group 4 (NumberOfMajorSurgeries == 3)')
```

**INFERENCE:**

- From **QQ Plots**, **Shapiro-Wilk's Test** it is clear that the one of the group of **Data is Not Normal**.

- Let us check for **Equal Variance** in Each Group- **Levene's Test**

**LEVENE'S TEST**


```python
sns.histplot(data = df, x = df['PremiumPrice'], hue = df['NumberOfMajorSurgeries'])
```

```python
# LEVENE TEST TO CHECK EQUAL VARIANCE

# H0: All groups have the Same Variance
# Ha: One or more groups have Different Variance

from scipy.stats import levene

# Perform Levene's test for homogeneity of variances across the groups
test_stat, p_value = levene(group1, group2, group3, group4)
print('p_value : ', p_value)
print('Test Statistic:', test_stat)

# Alpha level
alpha = 0.05

# Hypothesis test conclusion
if p_value < alpha:
    print("Reject H0, implies variance is different across the groups")
else:
    print("Fail to reject H0, All groups have the same variance")
```

**INFERENCE:**

- The results from **Levene's Test** indicate that the assumption of **Equal Variance** does not hold true.
- As a result, the assumptions for **ANOVA are not satisfied**.

Therefore, the **Kruskal-Wallis Test** should be the appropriate **Non Parametric** test to apply.

Since the **assumption of equal variance** among the groups is **violated**, the results from **ANOVA** may not be reliable. To address this issue and ensure the validity of the analysis, we can **perform the Kruskal-Wallis test**, which serves as a **non-parametric alternative to ANOVA**.

**KRUSKAL-WALLIS TEST**

```python
from scipy.stats import kruskal

#H0 : There is no significant difference in Premium Price across different levels of NumberOfMajorSurgeries
#Ha : There is a significant difference in Premium Price across different levels of NumberOfMajorSurgeries

# Perform Kruskal-Wallis test
kruskal_stat, p_value = kruskal(group1, group2, group3, group4)

# Display the results
print("Kruskal-Wallis Test Statistic:", kruskal_stat)
print("p-value:", p_value)

# For alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in Premium Price across different levels of NumberOfMajorSurgeries.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in Premium Price across different levels of NumberOfMajorSurgeries.")
```

**INFERENCE:**

Since the **p-value is less than the alpha**, we **reject the null hypothesis**. Therefore, the **Kruskal-Wallis test** **confirms** that there is a **significant difference** in **Premium Price** across the different levels of **NumberOfMajorSurgeries**.

## **CHI-SQUARE TEST**

 - Categorical Vs Categorical Features

### **Assumptions of the Chi-Square Test**  

1. **Categorical Data**  
   - Both variables must be **categorical** (e.g., Age Group & Health Status).  

2. **Independence of Observations**  
   - Each observation should be **independent**, meaning no repeated measures or paired data.  

3. **Adequate Sample Size**  
   - The expected frequency for **each cell** in the contingency table should be ** 5** for the test to be valid.  
   - If any expected count is **< 5**, consider using **Fishers Exact Test** instead.  

4. **Mutually Exclusive Categories**  
   - Each observation must fit into **only one** category of each variable (no overlap).  

5. **Random Sampling**  
   - The sample should be randomly selected to generalize results properly.  

### **5. Is there a significant relationship between Age Group and Health Status?**



To test whether **Age Group** and **Health Status** are dependent (i.e., there's a relationship between them), you can frame the hypothesis as follows:

 - **Null Hypothesis (H):** Age Group and Health Status are independent.  

- **Alternative Hypothesis (H)**: Age Group and Health Status are dependent.


```python
from scipy.stats import chi2_contingency

# Create a contingency table for AgeGroup vs HealthStatus
contingency_table = pd.crosstab(df['AgeGroup'], df['HealthStatus'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Display results
print("Contingency Table:")
print(contingency_table)
print("\nExpected Frequencies:")
print(expected)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)

# Check if expected frequencies are greater than or equal to 5
if (expected >= 5).all():
    print("\nThe expected frequencies are sufficient for the Chi-Square test.")
else:
    print("\nSome expected frequencies are less than 5, Chi-Square test may not be valid.")
```

**INFERENCE:**

- The results indicate that the **expected frequencies** for each cell in the contingency table are ** 5**, which satisfies the conditions for the **Chi-Square test to be valid**.

- Now, let's **interpret the p-value** to determine whether there is a significant relationship between **Age Group** and **Health Status**.

This will help us assess whether these two groups are dependent or not.









```python
# Interpretation
alpha = 0.05
if p_value < alpha:
    print('Reject the null hypothesis : Age Group and Health Status are Dependent.')
else:
    print('Fail to reject the null hypothesis : Age Group and Health Status are Independent.')
```

**INFERENCE:**

- The interpretation of the **p-value** clearly indicates that **Age Group** and **Health Status** are **dependent** on each other.

### **6. Is there a significant relationship between categorical variables with only two unique values in the dataset?**

- **Alternative Hypothesis (H):** There is a **significant** association between the **binary categorical variables** (i.e., they are **Dependent**)

```python
# Creating a new List for binary categorical Variable

binary_cat_cols = [col for col in df.columns if df[col].nunique() == 2]
binary_cat_df = df[binary_cat_cols]
print(binary_cat_cols)
```

```python

from scipy.stats import chi2_contingency

# Create a dictionary to store p-values
p_values = {}

# Iterate through all combinations of binary categorical columns
for col1 in binary_cat_cols:
    p_values[col1] = {}
    for col2 in binary_cat_cols:
        if col1 != col2:
            # Create a contingency table
            contingency_table = pd.crosstab(binary_cat_df[col1], binary_cat_df[col2])

            # Perform chi-square test
            chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

            # Store p-value
            p_values[col1][col2] = p_value

            # Display results
            print("Contingency Table:")
            print(contingency_table)
            print("\nExpected Frequencies:")
            print(expected)
            print("\nChi-Square Statistic:", chi2_stat)
            print("Degrees of Freedom:", dof)
            print("P-value:", p_value)

            alpha = 0.05
            if p_value < alpha:
              print('Reject Null Hypothesis : There is a significant association between the binary categorical variables (i.e., they are Dependent')
            else:
              print('Fail to Reject Null Hypothesis : There is no significant association between the binary categorical variables (i.e., they are Independent')



```
